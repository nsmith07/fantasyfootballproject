```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('fantasyfootball.csv')
```

```python
df2 = pd.read_csv('fantasyfootballrank.csv')
```

```python
finaldf = pd.merge(df, df2, on = "Player", how = "inner")
```

```python
finaldf = finaldf.rename(columns={'Tm': 'Team', 'FantPos': 'Pos', 'Yds': 'PassingYds','TD.3': 'AllTD', 'Att': 'PassAtt', 'Yds.2': 'RecYds', 'TD.2': 'RecTD','TD.1':'RushTD','TD':'PassTD','Att.1': 'RushAtt', 'Yd.1':'RushYds'})
finaldf.head()
```

```python
' , '.join(finaldf.columns)
```

```python

finaldf = finaldf.rename(columns={'PPR': 'FantasyPoints', 'AVG': 'AverageDraftPosition'})
```

```python
finaldf.head()
```

```python
rbdf = finaldf[finaldf['Pos'] == 'RB']
qbdf = finaldf[finaldf['Pos'] == 'QB']
wrdf = finaldf[finaldf['Pos'] == 'WR']
tedf = finaldf[finaldf['Pos'] == 'TE']
```

```python
rbdf.head()
```

```python
rush_column = ['RushAtt', 'Yds.1', 'Y/A', 'RushTD', 'AllTD']
recieving_column = ['Tgt', 'Rec', 'RecYds', 'Y/R', 'RecTD']
pass_column = ['Cmp', 'PassAtt', 'PassingYds', 'PassTD', 'Int']
```

```python
def transform_columns(df, new_column):
    df = df[['Player', 'Team', 'G', 'GS','FantasyPoints', 'Age', 'AverageDraftPosition'] + new_column]
    return df

rbdf = transform_columns(rbdf, rush_column+recieving_column)
wrdf = transform_columns(wrdf, recieving_column+rush_column)
qbdf = transform_columns(qbdf, pass_column+rush_column)
tedf = transform_columns(tedf, recieving_column+rush_column)
```

```python
rbdf = rbdf.rename(columns={'Yds.1': 'RushYds'})
rbdf.head()
```

```python
wrdf = wrdf.rename(columns={'Yds.1': 'RushYds'})
tedf = tedf.rename(columns={'Yds.1': 'RushYds'})
qbdf = qbdf.rename(columns={'Yds.1': 'RushYds'})
```

```python
qbdf.head()
```

```python
rbdf['FantasyPoints/GM'] = rbdf['FantasyPoints']/rbdf['G']
rbdf['FantasyPoints/GM'] = rbdf['FantasyPoints/GM'].apply(lambda x: round(x, 2))

rbdf['Usage/GM'] = (rbdf['RushAtt'] + rbdf['Tgt'])/rbdf['G']
rbdf['Usage/GM'] = rbdf['Usage/GM'].apply(lambda x: round(x, 2))


wrdf['FantasyPoints/GM'] = wrdf['FantasyPoints']/wrdf['G']
wrdf['FantasyPoints/GM'] = wrdf['FantasyPoints/GM'].apply(lambda x: round(x, 2))

wrdf['Usage/GM'] = (wrdf['RushAtt'] + wrdf['Tgt'])/wrdf['G']
wrdf['Usage/GM'] = wrdf['Usage/GM'].apply(lambda x: round(x, 2))

tedf['FantasyPoints/GM'] = tedf['FantasyPoints']/tedf['G']
tedf['FantasyPoints/GM'] = tedf['FantasyPoints/GM'].apply(lambda x: round(x, 2))

tedf['Usage/GM'] = (tedf['RushAtt'] + tedf['Tgt'])/tedf['G']
tedf['Usage/GM'] = tedf['Usage/GM'].apply(lambda x: round(x, 2))
```
